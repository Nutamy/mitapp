4. Во многих спам письмах вы можете увидеть слово ```"urgent"```. Однако некоторые электронные письма не являются спамом, даже если они содержат слово «urgent». Допустим, мы знаем следующие вероятности:

    * Вероятность получить **спам письмо** по электронной почте: ```0.27```
    
    * Вероятность того, что электронное **спам** письмо содержит слово ```"urgent"```: ```0.46```
    
    * Вероятность того, что электронное **НЕ спам** письмо содержит слово ```"urgent"```: ```0.19```
    
    

$P(HIV^+|Test^+)$ = $P(HIV^+ \cap Test^+)/P(HIV^+)$

$P(Test^+|HIV^+)$ = $P(HIV^+ \cap Test^+)/P(Test^+)$

$P(HIV^+|Test^+)$ = $(P(HIV^+) * P(Test^+|HIV^+))/P(Test^+)$

Пожалуйста, рассчитайте следующие вероятности:

* $P("spam")$, и присвойте результат переменной ```p_spam```.
* $P("non-spam")$, и присвойте результат переменной ```p_non_spam```.
* $P("urgent"|"spam")$, и присвойте результат переменной ```p_urgent_given_spam```.
* $P("urgent"|"non-spam")$, и присвойте результат переменной ```p_urgent_given_non_spam```.
* $P("urgent"\cap "spam")$, и присвойте результат переменной ```p_urgent_and_spam```.
* $P("urgent"\cap "non-spam")$, и присвойте результат переменной ```p_urgent_and_non_spam```.
* $P("urgent")$, и присвойте результат переменной ```p_urgent```.
* $P("spam"|"urgent")$, и присвойте результат переменной ```p_spam_given_urgent```.
* $P("non-spam"|"urgent")$, и присвойте результат переменной ```p_non_spam_given_urgent```.


Формула полной вероятности:
$P(d) = P(B)P(d|B) + P(A)P(d|A)$

$P(HIV^+|Test^+)$ = $P(HIV^+ \cap Test^+)/P(Test^+)$

$P(HIV^+|Test^+)$ = $P(HIV^+)P(Test^+|HIV^+)/P(Test^+)$

$P(HIV^+ \cap Test^+) = P(HIV^+)P(HIV^+|Test^+)$

````python
#Начало вашего кода
p_spam = 0.27
p_non_spam = 1 - p_spam
p_urgent_given_spam = 0.46
p_urgent_given_non_spam = 0.19
p_urgent_and_spam = p_spam * p_urgent_given_spam
p_urgent_and_non_spam = p_non_spam * p_urgent_given_non_spam
p_urgent = p_spam * p_urgent_given_spam + p_non_spam * p_urgent_given_non_spam
p_spam_given_urgent = p_urgent_and_spam/p_urgent
p_non_spam_given_urgent = 1 - p_spam_given_urgent
#Конец вашего кода

print(f"{p_spam*100:.2f}%")
print(f"{p_non_spam*100:.2f}%")
print(f"{p_urgent_given_spam*100:.2f}%")
print(f"{p_urgent_given_non_spam*100:.2f}%")
print(f"{p_urgent_and_spam*100:.2f}%")
print(f"{p_urgent_and_non_spam*100:.2f}%")
print(f"{p_urgent*100:.2f}%")
print(f"{p_spam_given_urgent*100:.2f}%")
print(f"{p_non_spam_given_urgent*100:.2f}%")
````

    * Присвойте априорную вероятность получить спам по электронной почте переменной: ```prior```.
    
    * Присвойте апостериорную вероятность получить спам по электронной почте переменной: ```posterior```.
    
    * Рассчитайте соотношение между апостериорной и априорной вероятностью - вам нужно разделить апостериорную вероятность на априорную вероятность. Присвойте ответ переменной: ```ratio```.

````python
#Начало вашего кода
prior = p_spam
posterior = p_spam_given_urgent
ratio = posterior/prior

'''
априорная вероятность - это вероятность получить событие не имея никаких дополнительных данных или условий
вероятность правильно предсказать, что письмо спам 27%

апостериорная вероятность - это вероятность события с учётом того, что у нас есть уточняющая информация
есть ещё одно событие, которое уже произошло
так вероятность того, что полученное письмо спам, если оно содержит слово 'urgent' 47.24%

Соотношение между этими вероятностями показывает насколько сильно одно событие влияет на другое

Например априорная вероятность лететь на Боинге 75%
а апостприорная вероятность того, что вы летите на Боинге, если произошло событие - опоздание 85.7%
85.7/75 = 1.14

Я понимаю это так, что опоздание не так сильно влияет на событие лететь на Боинге, 
как событие 'urgent' влияет на событие - получить спам

Так получение письма со словом 'urgent' увеличивает шансы получить спам в 1.75 раза
'''
#Конец вашего кода
print(f"{prior*100:.2f}%")
print(f"{posterior*100:.2f}%")
print(f"{ratio:.2f}")
````
