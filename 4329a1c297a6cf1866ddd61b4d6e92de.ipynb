3. Пожалуйста, смоделируйте эксперименты с монетой. Предположим, что есть одинаковый шанс получить Heads и Tails после одного броска монеты. Оцените вероятность получения Tails, если вы подбрасываете монету от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:




````python
#Начало вашего кода
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.seed = 1

theoretical = [1/2 for x in range(10000)]
x = [x for x in range(10000)]

main_simulation = []

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot()

min_flips = 10
max_flips = 10000

for flip in range(min_flips, max_flips):
    simulation = []
    for f in range(flip):
        simulation.append(np.random.choice([0,1]))
    p_flip = sum(simulation)/flip
    main_simulation.append(p_flip)

ax.plot(main_simulation, color='#1F77B4', ls=':', label='Simulation')
ax.plot(x,theoretical, color='red', ls=':', lw=2, label='Theoretical')

ax.set_title('P(Tails)')
ax.set_ylabel('Probability of Tails')
ax.set_xlabel('# of flips')
ax.legend()
ax.set_xlim(0,10000)
ax.set_ylim(0,1)
#Конец вашего кода
````

4. Пожалуйста, смоделируйте эксперименты с игральной костью (со сторонами ```{1,2,3,4,5,6}```). Предположим, что есть равный шанс получить любую из шести сторон после броска игральной кости. Оцените вероятность получить число меньше или равное 2 после броска игральной кости от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:



````python
#Начало вашего кода
theoretical = [2/6 for x in range(10000)]
x = [x for x in range(10000)]

main_simulation = []

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot()

min_flips = 10
max_flips = 10000

for flip in range(min_flips, max_flips):
    simulation = []
    for f in range(flip):
        dice_roll = np.random.choice([1,2,3,4,5,6])
        if dice_roll <= 2:
            simulation.append(1)
    p_flip = sum(simulation)/flip
    main_simulation.append(p_flip)

ax.plot(main_simulation, color='#1F77B4', ls=':', label='Simulation')
ax.plot(x,theoretical, color='red', ls=':', lw=2, label='Theoretical')

ax.set_title('P(<3)')
ax.set_ylabel('Probability')
ax.set_xlabel('# of flips')
ax.legend()
ax.set_xlim(0,10000)
ax.set_ylim(0,1)
#Конец вашего кода
````

5. Пожалуйста, смоделируйте эксперименты с игральной костью (со сторонами ```{1,2,2,3,4,5}```). Предположим, что есть равный шанс получить любую из шести сторон после броска игральной кости. Оцените вероятность получить число меньше или равное 3 после броска игральной кости от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:



````python
#Начало вашего кода
theoretical = [4/6 for x in range(10000)]
x = [x for x in range(10000)]

main_simulation = []

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot()

min_flips = 10
max_flips = 10000

for flip in range(min_flips, max_flips):
    simulation = []
    for f in range(flip):
        dice_roll = np.random.choice([1,2,2,3,4,5])
        if dice_roll <= 3:
            simulation.append(1)
    p_flip = sum(simulation)/flip
    main_simulation.append(p_flip)

ax.plot(main_simulation, color='#1F77B4', ls=':', label='Simulation')
ax.plot(x,theoretical, color='red', ls=':', lw=2, label='Theoretical')

ax.set_title('P(<4)')
ax.set_ylabel('Probability of getting number < 4')
ax.set_xlabel('# of flips')
ax.legend()
ax.set_xlim(0,10000)
ax.set_ylim(0,1)
#Конец вашего кода
````
