````python
def print_new_and_old_values(data, column):
    uniq = sorted(data[column].unique(), reverse=False)
    print(f'column = {column}')
    for new,old in enumerate(uniq):
        print(f'new = {new}')
        print(f'old = {old}')
        print('-------------')
    print('\n')
        
print_new_and_old_values(exercise, 'diet')
print_new_and_old_values(exercise, 'time')
print_new_and_old_values(exercise, 'kind')
````

````python
pd_data=exercise.copy()
# Начало вашего кода
def convert_from_str_to_int(data, column):
    new_column = f'{column}-int'
    data[new_column] = data[column]
    uniq = sorted(data[column].unique(), reverse=False)

    for int_value, str_value in enumerate(uniq):
        data[new_column] = data[new_column].replace(str_value, int_value)

    data[new_column] = data[new_column].astype('int64')
    
convert_from_str_to_int(pd_data, 'diet')
convert_from_str_to_int(pd_data, 'time')
convert_from_str_to_int(pd_data, 'kind')
pd_data = pd_data.drop('Unnamed: 0', axis=1)

# Конец вашего кода
pd_data.head()
````
