### Шаг 4

Пожалуйста, реализуйте функцию ```plot_sns_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_pclass(data=None):
    #Начало вашего кода
    #изменил ширину чтобы лучше отображалось
    for idx,v in enumerate(["First","Second","Third"],1):
        sns.set_context("talk");
        data1=data[data['class']==v].sort_values(ascending=False,by=['alive'])
        g=sns.relplot(x="fare",y="age",hue="alive",data=data1);
        g.fig.set_figheight(4)
        g.fig.set_figwidth(10)
        g.fig.suptitle(f"Passenger class: {idx}");
    #Конец вашего кода

plot_sns_pclass(data=titanic)
````

### Шаг 5


Пожалуйста, реализуйте функцию ```plot_sns_sex(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе пола пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_sex(data=None):
    #Начало вашего кода
    for idx,v in enumerate(data["sex"].value_counts(ascending=True).index.to_list(),1):
        sns.set_context("talk");
        data1=data[data['sex']==v].sort_values(ascending=False,by=['alive'])
        g=sns.relplot(x="fare",y="age",hue="alive",data=data1);
        g.fig.set_figheight(4)
        g.fig.set_figwidth(12)
        g.fig.suptitle(f"Passenger sex: {v}");
    #Конец вашего кода

plot_sns_sex(data=titanic)
````

### Шаг 6


Пожалуйста, реализуйте функцию ```plot_sns_town(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе города пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_town(data=None):
    #Начало вашего кода
    for idx,v in enumerate(["Cherbourg","Queenstown","Southampton"],1):
        sns.set_context("talk");
        data1=data[data['embark_town']==v].sort_values(ascending=False,by=['alive'])
        g=sns.relplot(x="fare",y="age",hue="alive",data=data1);
        g.fig.set_figheight(4)
        g.fig.set_figwidth(12)
        g.fig.suptitle(f"Passenger town: {v}");
    #Конец вашего кода

plot_sns_town(data=titanic)
````
