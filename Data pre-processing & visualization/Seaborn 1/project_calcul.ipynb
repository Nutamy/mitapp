```python
import sys
def arithmetic_arranger(list_of_problems):
    def solving(task,s):
        a,b = task.split(s)
        lenth = len(a)+1 if len(a)>len(b) else len(b)+1
        try:
            a,b = int(a),int(b)
            line_1 = ' '*(lenth-len(str(a)))+str(a)+' '*4
            line_2 = s + ' '*(lenth-len(str(b))-1)+str(b)+' '*4
            line_3 = '-'*lenth+' '*4
            if s == '-':
                c = int(a) - int(b)
            else:
                c = int(a) + int(b)
            line_4 = ' '*(lenth-len(str(c)))+str(c)+' '*4
            return line_1, line_2, line_3, line_4
        except:
            print('Error: Numbers must only contain digits.')
            sys.exit()

   
    line_1 = ''
    line_2 = ''
    line_3 = ''
    line_4 = ''

    for problem in list_of_problems:
        if '+' in problem:
            l1,l2,l3,l4 = solving(task=problem,s='+')
            line_1 += l1
            line_2 += l2
            line_3 += l3
            line_4 += l4
        elif '-' in problem:
            l1,l2,l3,l4 = solving(task=problem,s='-')
            line_1 += l1
            line_2 += l2
            line_3 += l3
            line_4 += l4
        else:
            print("Error: Operator must be '+' or '-'")

    if len(list_of_problems) > 4:
        print('Error: Too many problems.')
    else:
        print(f'{line_1}\n{line_2}\n{line_3}\n{line_4}')
    
arithmetic_arranger(["32 + 6a98", "3801 - 2", "45 + 43", "123 + 49", "123 + 49"])
print()
arithmetic_arranger(["32 + 8", "1 - 3801", "9999 + 9999", "523 - 49"])
```
