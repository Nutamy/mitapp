5. Напишите функцию, которая проверяет, равен ли какой-либо из элементов данного массива numpy нулю.

````python
arr = np_arr_rand
def check_zeros(input_arr):
    # Начало вашего кода
    return np.any(input_arr==0)
    # Конец вашего кода
print(check_zeros(arr))
````

### Ожидаемый результат:

