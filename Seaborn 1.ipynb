### Шаг 4

Пожалуйста, реализуйте функцию ```plot_sns_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_pclass(data=None):
    sns.reset_orig()
    #Подготовка данных
    x='fare'
    y='age'
    hue='alive'
    pclasses = sorted(titanic['pclass'].unique())
    unique_alive = sorted(titanic['alive'].unique().tolist(), reverse=True)
    
    #Построение графиков
    for pclass in pclasses:
        fig = plt.figure(figsize=(8,4))
        ax = fig.add_subplot()
        sns.scatterplot(x=x, y=y, hue=hue, hue_order=unique_alive, data=data[data['pclass']==pclass]).set(title=f'Passenger class: {pclass}')
        sns.despine(trim=False)
    
        #Работа со стилями
        sns.set_context('poster')
        sns.set(style='white')
    
    plt.subplots_adjust(wspace=0, hspace=0.5)
    #Конец вашего кода

plot_sns_pclass(data=titanic)
````

### Шаг 5


Пожалуйста, реализуйте функцию ```plot_sns_sex(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе пола пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_sex(data=None):
    #Подготовка данных
    x='fare'
    y='age'
    hue='alive'
    col_sex = sorted(titanic['sex'].unique())
    unique_alive = sorted(titanic['alive'].unique().tolist(), reverse=True)
    
    #Построение графиков
    for sex in col_sex:
        fig = plt.figure(figsize=(8,4))
        ax = fig.add_subplot()
        sns.scatterplot(x=x, y=y, hue=hue, hue_order=unique_alive, data=data[data['sex']==sex]).set(title=f'Passenger sex: {sex}')
        sns.despine(trim=False)
    
        #Работа со стилями
        sns.set_context('poster')
        sns.set(style='white')
    
    plt.subplots_adjust(wspace=0, hspace=0.5)

plot_sns_sex(data=titanic)
````

### Шаг 6


Пожалуйста, реализуйте функцию ```plot_sns_town(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе города пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_town(data=None):

    #Подготовка данных
    x='fare'
    y='age'
    hue='alive'
    towns = sorted(titanic['embark_town'].dropna().unique())
    unique_alive = sorted(titanic['alive'].unique().tolist(), reverse=True)
    
    #Построение графиков
    for town in towns:
        fig = plt.figure(figsize=(8,4))
        ax = fig.add_subplot()
        sns.scatterplot(x=x, y=y, hue=hue, hue_order=unique_alive, data=data[data['embark_town']==town]).set(title=f'Passenger town: {town}')
        sns.despine(trim=False)
    
        #Работа со стилями
        sns.set_context('poster')
        sns.set(style='white')
    
    plt.subplots_adjust(wspace=0, hspace=0.5)
    
    #Конец вашего кода

plot_sns_town(data=titanic)
````
